```python
features=['성별', '만나이', '월평균수입', '현재결혼상태', '아버지고혈압진단여부',
       '어머니고혈압진단여부', '아버지당뇨병진단여부', '어머니당뇨병진단여부', '음주여부', '총음주기간', '흡연여부',
       '총흡연기간(년)', '하루흡연량(개비)', '간접흡연여부', '규칙적운동여부(몸에땀이날정도)','폐경여부',
       '임신경험여부', '허리둘레(cm)', '엉덩이둘레(cm)', '신장(cm)','체중(kg)']

targets=[ '당뇨병진단여부']
```

```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc
%matplotlib inline
rc('font', family='AppleGothic')
matplotlib.rcParams['axes.unicode_minus'] = False    
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import KFold, GridSearchCV
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[features],df[targets],test_size=1/3,random_state=42, stratify=df[targets])
X_train,X_val,y_train,y_val = train_test_split(X_train,y_train,test_size=1/3,random_state=42, stratify=y_train)
```

```python
smote=SMOTE(k_neighbors=5)
smoted_X_train,smoted_y_train=smote.fit_resample(X_train,y_train)
smoted_X_train
```

```python
lgbm = LGBMClassifier(n_estimators=200)
params = {'max_depth': [10, 15, 20],
          'min_child_samples': [20, 40, 60],
          'subsample': [0.8, 1]}

grid = GridSearchCV(lgbm, param_grid=params)
grid.fit(smoted_X_train, smoted_y_train, early_stopping_rounds=100, eval_metric='auc',
         eval_set=[(smoted_X_train, smoted_y_train), (X_val, y_val)])

print("최적 파라미터: ", grid.best_params_)
lgbm_roc_score = roc_auc_score(y_test, grid.predict_proba(X_test)[:, 1], average='macro')
print("ROC AUC: {0:.4f}".format(lgbm_roc_score))

# 위 결과를 적용하여 재학습
lgbm = LGBMClassifier(n_estimators=1000, num_leaves=50, subsample=0.8,
                      min_child_samples=60, max_depth=20)

evals = [(X_test, y_test)]
lgbm.fit(smoted_X_train, smoted_y_train, early_stopping_rounds=100, eval_metric='auc',
         eval_set=evals, verbose=True)

score = roc_auc_score(y_test, grid.predict_proba(X_test)[:, 1], average='macro')
print("ROC AUC: {0:.4f}".format(score))
```
