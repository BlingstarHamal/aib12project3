## 결측치 조회
66666=조사안함,77777=해당없음,99999=미상/무응답/미측정

```python
# 결측치(val)가 포함된 열 찾는 함수
def nandata(data,val):
    col=[]
    for x in data.columns:
        if data[data[x]==val][x].count() > 0:
            col.append(x)
    print(f'{val}을 포함하는 열의 개수 {len(col)}')
    return col
```

```python
# 특정 컬럼의 결측치 확인
def nandf2(data,col):
    nanlist=pd.DataFrame()
    nannum=['66666','77777','99999']
    nanlist['결측값']=nannum
    nanlist=nanlist.set_index(keys='결측값')

    for x in data[col]:
        col=[]
        col.append(data[data[x]==66666][x].count())
        col.append(data[data[x]==77777][x].count())
        col.append(data[data[x]==99999][x].count())
        if sum(col) > 0:
            nanlist[x]=col
    print(nanlist.columns)
    return nanlist

#nanlist=nandf(data)
#nanlist
```

```python
# 결측치(66666,99999)를 평균값(0) 또는 최빈값(1)으로 대체하는 함수
def nanchange(col,am):
    nan_index=data[(data[col]==99999) | (data[col]==66666)].index
    
    if am == 0:
        x=round(np.mean(data[col].drop(nan_index)))
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 평균값(반올림) :',x)
    elif am == 1:
        x=int(mode(data[col].drop(nan_index))[0])
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 최빈값 :',x)
    else :
        print('평균값을 원하면 am=0, 최빈값을 원하면 am=1 입력해주세요')
        
    return data
```

```python
# 결측치(66666,99999)를 평균값(0) 또는 최빈값(1)으로 대체하는 함수 / 7777도 제외
def nanchange2(col,am):
    nan_index=data[(data[col]==99999) | (data[col]==66666) | (data[col]==77777)].index
    
    if am == 0:
        x=round(np.mean(data[col].drop(nan_index)))
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 평균값(반올림) :',x)
    elif am == 1:
        x=int(mode(data[col].drop(nan_index))[0])
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 최빈값 :',x)
    else :
        print('평균값을 원하면 am=0, 최빈값을 원하면 am=1 입력해주세요')
        
    return data
    
    
```

```python
# 결측치(66666,99999)를 평균값(0) 또는 최빈값(1)으로 대체하는 함수 / 99999만
def nanchange3(col,am):
    nan_index=data[(data[col]==99999)].index
    
    if am == 0:
        x=round(np.mean(data[col].drop(nan_index)))
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 평균값(반올림) :',x)
    elif am == 1:
        x=int(mode(data[col].drop(nan_index))[0])
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 최빈값 :',x)
    else :
        print('평균값을 원하면 am=0, 최빈값을 원하면 am=1 입력해주세요')
        
    return data
    
    
```

```python
# 결측치(66666,99999)를 평균값(0) 또는 최빈값(1)으로 대체하는 함수 / 평균 반올림x
def nanchange4(col,am):
    nan_index=data[(data[col]==99999) | (data[col]==66666)].index
    
    if am == 0:
        x=round(np.median(data[col].drop(nan_index)),1)
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 평균값(소수점첫째) :',x)
    elif am == 1:
        x=int(mode(data[col].drop(nan_index))[0])
        data[col]=data[col].replace(99999,x)
        data[col]=data[col].replace(66666,x)
        print(f'{col} 결측치 대체 최빈값 :',x)
    else :
        print('평균값을 원하면 am=0, 최빈값을 원하면 am=1 입력해주세요')
        
    return data
```

### 진단 결측치

```python
def sick_check(col_sick,col_age):
    data=nanchange(col_sick,1)
    index1=data[data[col_sick]==1][col_age].index
    for x in index1:
        data[col_age][x]=77777
        
    index2=data[(data[col_age]==99999) | (data[col_age]==66666) | (data[col_age]==77777)].index
    x=round(np.mean(data[col_age].drop(index2)))
    data[col_age]=data[col_age].replace(99999,x)
    data[col_age]=data[col_age].replace(66666,x)
    print(f'{col_age} 결측치 대체 평균값(반올림) :',x)
    
    return data
```
